```python
certified,denied=[],[]
for i in continents:
    con1=(visadataset['continent']==i)
    con2=(visadataset['case_status']=='Denied')
    con3=(visadataset['case_status']=='Certified')
    count_denied=len(visadataset[con1&con2])  
    count_certified=len(visadataset[con1&con3])
    certified.append(count_certified)
    denied.append(count_denied)
```
