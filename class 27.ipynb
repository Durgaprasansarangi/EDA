***Pearson-correlation-coefficient(r)***

```python
- 100  to 100

r=-1  to   1


-1  to 0 ======== > negative relation

-0.5 : both fetaures 50% negatively correlated
    
0 to 1 ========= > postively related
    
0.7: both features 70% postively correlated
    
0 : No relation/ independent each other
```

```python
r will vary  -1 to 1

-1 to 0 ====== negative

0 to 1 ======= post

=
```

***difference between corelation and covariance***

```python
Covariance:  will exalain wether the both features have relation or not
    
Correlation: will explain how much the relation between two features
```
