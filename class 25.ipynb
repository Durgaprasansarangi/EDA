```python
Q1=round(np.quantile(data,0.25),2)
Q2=round(np.quantile(data,0.50),2)
Q3=round(np.quantile(data,0.75),2)
IQR=(Q3-Q1)
LB=Q1-1.5*IQR
UB=Q3+1.5*IQR

print('Q1:',Q1)
print('Q2:',Q2)
print('Q3:',Q3)
print('IQR:',IQR)
print('LB:',LB)
print('UB:',UB)
```

```python
data=visadataset['prevailing_wage']
con1=data<LB
con2=data>UB
con1|con2
outliers_data=visadataset[con1|con2]
#visadataset[(data>7982.34003)&
            #(data<15962.5434)]
```
